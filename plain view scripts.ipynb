```python
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime
%matplotlib inline
```

```python
filename  = 'GeostrophicVel_0009-01-06.nc'
dataset = Dataset(filename)
dataset.variables.keys()
```

```python
ugos = np.array(dataset.variables['UGOS'])
vgos = np.array(dataset.variables['VGOS'])
```

```python
velmag = np.sqrt(ugos*ugos + vgos*vgos)
```

```python
fig = plt.figure(figsize=(15,10))
print(np.shape(ugos))
plt.pcolormesh(ugos[0,:,:])
plt.colorbar()
```

```python
fig = plt.figure(figsize=(15,10))
print(np.shape(ugos))
plt.pcolormesh(velmag[0,:,:])
plt.clim(0,80)
plt.colorbar()
```
