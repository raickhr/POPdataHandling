## Conversion terms

### Mean Available Potential Energy to Mean Kinetic Energy
$$C(P_m,K_m) = - \int_V g {\overline{\rho}}\,{\overline{w}} dV $$

### Eddy Available Potential Energy to Eddy Kinetic Energy
$$C(P_e,K_e) = - \int_V g \overline{\rho' w'} dV$$

### Eddy Kinetic Energy to Mean Kinetic Energy

$$C(K_e,K_m) = \int_V (\rho_o \overline{u'\boldsymbol{u'}}  \nabla \overline{u} + \rho_o \overline{v'\boldsymbol{u'}}. \nabla\overline{v}) dV $$

### Eddy Available Potential Energy to Mean Available Potential Energy

$$C(P_e,P_m) = - \int_V \frac{g}{n_o} \overline{\rho'\boldsymbol{u'}}. \nabla_h \overline{\rho} dV$$


## Generation Terms

### Generation of Mean Available Potential Energy

$$ G(P_m) = - \int_S g \frac{\alpha_{o,1}}{n_o} \overline{J}_s \overline{\rho*} dS  - \int_S g \frac{\beta_{o,1}}{n_o} \overline{G}_s \overline{\rho*} dS$$

here J is surface temperature flux<br>
G is surface salinity flux <br>
$\alpha_{o,1}$ and $\beta_{o,1} $ are coefficient of expansions at the top layer<br>

$J_s = \frac{H}{\rho_s c} $ and H is surface heat flux <br>
$G_s = S_1(E - P) $ <br>
E is evaporation and P is precipitation

$S_1$ is the surface salinity </br>

$\rho* = \rho - \rho_{ref}$ <br>
$\rho_{ref} = \langle\overline{\rho}\rangle $ where $\langle ... \rangle $ refer to area averaging <br>

### Generation of mean Kinetic Energy

$$G(K_m) = \int_S(\overline{\tau_x}\,\overline{u} + \overline{\tau_y}\,\overline{v})\, dS $$

### Generation of Eddy Kinetic Energy

$$G(K_e) = \int_S(\overline{\tau_x' \,u'} + \overline{\tau_y' \,v'})\, dS $$

### Generation of Eddy Available Potential Energy
$$ G(P_e) = - \int_S g \frac{\alpha_{o,1}}{n_o} \overline{J_s' \rho'} dS  - \int_S g \frac{\beta_{o,1}}{n_o} \overline{G_s' \rho'} dS$$


## Energy reservoirs

### Mean Potential Energy

$$ P_m = - \int_V \frac{g}{2n_0}\overline{\rho*}^2 dV $$

### Mean Kinetic Energy

$$ K_m = \int_V \frac{1}{2} \rho_0 \left( \overline{u}^2 + \overline{v}^2 \right) $$

### Eddy Potential Energy
$$ P_e = - \int_V \frac{g}{2n_0}\overline{\rho*'^2} dV $$

### Eddy Kinetic Energy
$$ K_e = \int_V \frac{1}{2} \rho_0 \left( \overline{u'^2} + \overline{v'^2} \right) $$

## On the value of $n_o$ and Assumptions made for $\varrho,\; \alpha,\; and \;\beta$

$$ \frac{d\rho}{dt} = \left(\frac{\partial \rho }{\partial \theta}\right)_{p,S} 
\frac{d\theta}{dt} + \left(\frac{\partial \rho}{\partial S}\right)_{p,\theta}\frac{dS}{dt} + \left(\frac{\partial \rho}{\partial z}\right)_{\theta,S} w $$

$$ \frac{d \theta}{dt} = \frac{\partial J}{\partial z}  \;  and \; \frac{dS}{dt} = \frac{\partial G}{\partial z} $$

This gives us <br>

$$ \frac{\partial \rho}{\partial t} +
   \boldsymbol{u_h} . \nabla_h \rho  +
   w \left( \frac{\partial \rho }{\partial z} - 
   \left( \frac{\partial \rho }{\partial z}\right)_{S,\theta} \right)
   = \left(\frac{\partial \rho}{\partial \theta} \right)_{S,p} \frac{\partial J}{\partial z}  +
   \left(\frac{\partial \rho }{\partial S} \right)_{S, \theta} \frac{\partial G}{\partial z}
$$ 

$$ \frac{\partial \rho}{\partial t} +
   \boldsymbol{u_h} . \nabla_h \rho  +
   w \left( \frac{\partial \varrho}{\partial z} \right)
   = \alpha_o(\lambda,\phi,z) \frac{\partial J}{\partial z}  +
   \beta_o(\lambda,\phi,z) \frac{\partial G}{\partial z}
$$ 

$$\frac{\partial \rho }{\partial z} - 
   \left( \frac{\partial \rho }{\partial z}\right)_{S,\theta} = 
   \frac{\partial \varrho}{\partial z}  $$ is vertial local potential density gradient
   
$$ \frac{\partial \varrho}{\partial z} \approx \frac{\partial \langle \overline{\varrho} \rangle}{\partial z} = n_o(z) $$

$$ \left(\frac{\partial \rho}{\partial \theta} \right)_{S,p} \approx
\left(\frac{\partial \overline{\rho}}{\partial \theta} \right)_{S,p} = \alpha_o(\lambda,\phi,z)$$

$$ \left(\frac{\partial \rho}{\partial S} \right)_{\theta,p} \approx
\left(\frac{\partial \overline{\rho}}{\partial S} \right)_{\theta,p} = \beta_o(\lambda,\phi,z)$$

## The value of $\alpha$ and $\beta$ used in POP2 in linear Equation of State
$$ \alpha_o  = \left (\frac{\partial \rho}{\partial T}\right)_{S,p} = - 2.55 \times 10^{-4} gr/cm^3/K $$
$$ \beta_o = \left (\frac{\partial \rho}{\partial S}\right)_{T,p} = 7.64\times 10^-1 gr/cm^3/msu $$

$$ 1 gram/kg = 0.001 msu $$

# Main differences with the paper

- The paper has NCEP-NCAR reanalysis-1 forcing. (Reanalysis 1 has some error in the wind stress. Reanalysis -2 has corrected value).  I have used JRA-55 forcing. JRA-55 is second generation ocean product compared to 1st generation NCEP-NCAR 

- The grid is 0.1 deg tripolar but I have have 0.25 deg tripolar grid

- The model is spun up for 25 years and forced with OMIP, ERA-15 forcing [Read Roske 2006](http://citeseerx.ist.psu.edu/viewdoc/download?doi=10.1.1.518.6766&rep=rep1&type=pdf) for details of this forcing. I have started with a given initial value and used data of 9th year.

- The NCEP-NCAR forcing is 6th hourly for after spin up while I have to check this

- The fields are calculated from the instantenous values of every 600s but CESM has averaged from every time step. But I calculate the values of the fields from the 5 day averages of the output of CESM

- Evaporation and Precipitaion are only consided in the simulation in paper. CESM has full Run off river, snowfall, ice melt with evaporation and precipitation. But for now I have used only evaporation and precipitation.

- $\alpha_0$ and $\beta_0$ have spatial variations but I have used constant value for those. Also CESM setup may not be using the equation of State with constant $\alpha_0$ and $\beta_0$. Need to confirm this
